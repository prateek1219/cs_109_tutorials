```python
#Handling Duplicates
df2 = pd.DataFrame({'a': ['one', 'one', 'two', 'two', 'two', 'three', 'four'],
                 'b': ['x', 'y', 'x', 'y', 'x', 'x', 'x'],
                  'c': np.random.randn(7)})


#duplicated returns a boolean vector whose length is the number of rows, and which indicates whether a row is duplicated.
#drop_duplicates removes duplicate rows.


#By default, the first observed row of a duplicate set is considered unique, but each method 
#has a keep parameter to specify targets to be kept.

#keep='first' (default): mark / drop duplicates except for the first occurrence.
#keep='last': mark / drop duplicates except for the last occurrence.
#keep=False: mark / drop all duplicates.

df2.duplicated('a')
df2.duplicated('a', keep='last')
df2.duplicated('a', keep=False)

df2.drop_duplicates('a')
df2.drop_duplicates('a', keep='last')
df2.drop_duplicates('a', keep=False)

#you can check duplicacy in column tuples as well
df2.duplicated(['a','b']) #will check for (a,b) tuple duplcates Eg: if (a,b) = (one,x) appears again in any column
#then that col will be marked as duplicate

```

```python
#dropping duplicates indexes
df3 = pd.DataFrame({'a': np.arange(6),
                    'b': np.random.randn(6)},
                    index=['a', 'a', 'b', 'c', 'b', 'a'])
df3
l = df3.index.duplicated(keep='last') #you can mention keep=False/True as well
#this will give boolean array

#to remove dup index
df3[~l]

df3.get('a') #to get all values of column 'a'
#df3
```

```python
#Grouping

#By “group by” we are referring to a process involving one or more of the following steps

#Splitting the data into groups based on some criteria
#Applying a function to each group independently
#Combining the results into a data structure

df = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar',
                          'foo', 'bar', 'foo', 'foo'],
                 'B' : ['one', 'one', 'two', 'three',
                            'two', 'two', 'one', 'three'],
                   'C' : np.random.randn(8),
             'D' : np.random.randn(8)})
df

```

```python
df.groupby('A').sum() #group by A and then apply function=sum()
```

```python
df.groupby(['A','B']).sum() #group by A and on the subgroups (group by B) and then apply function=sum()
df.groupby(['A'], sort=False).sum() #here sort=False means that it will present the items in exact order as columns; 
#not sorted order
```

```python
grouped = df.groupby(['A', 'B'])

#grouped.groups
for name, group in grouped:
    print(name)
    print(group)

grouped.get_group(name=('bar','one')) 
#grouped.get_group('bar','one') #since you used [A,B] for grouping, you can give one more filter to get the group

#grouped.<TAB> to see the options available
```

```python
#Aggregate
#once you have grouped then you can apply variety of function on the groups
grouped.agg({'C' : np.sum, 'D' : lambda x: np.std(x, ddof=1)}) #diff func on diff columns
grouped.agg([np.sum, np.mean]) #on all columns of groups
grouped['C'].agg([np.sum, np.mean, np.std]) #on just one column of groups
```

```python
#Filtration
#applying filter on groups
dff = pd.DataFrame({'A': np.arange(8), 'B': list('aabbbbcc')})

dff.groupby('B').filter(lambda x: len(x) > 2)
dff.groupby('B').filter(lambda x: len(x) > 2, dropna=False)
```
